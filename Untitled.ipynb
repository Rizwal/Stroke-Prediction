```python
x=df.drop(['id','stroke','smoking_status'],axis=1)
y=df['stroke']
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import catboost as cb
from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
```

# Logistic Regression

```python
parameters={'max_iter':[100,200,300]}
lr=LogisticRegression()
lr_rs=RandomizedSearchCV(lr,param_distributions=parameters,cv=5,random_state=42)
lr_rs.fit(xtrain,ytrain)
```

```python
y_pred=lr_rs.predict(xtest)
```

```python
ac_sc=accuracy_score(ytest,y_pred)
ac_sc
```

# Random Forest

```python
parameters={'n_estimators':np.arange(20,100,10),'max_depth':np.arange(2,16,2)}
rf=RandomForestClassifier()
rf_rs=RandomizedSearchCV(rf,param_distributions=parameters,cv=5,random_state=42)
rf_rs.fit(xtrain,ytrain)
```

```python
y_pred=rf_rs.predict(xtest)
```

```python
ac_sc=accuracy_score(ytest,y_pred)
ac_sc
```

```python
dfnew=pd.DataFrame({'Actual':ytest,'Predicted':y_pred})
submit=dfnew.to_csv('submit.csv')
```
